# Dogecoin Price Prediction

> Predicting the price of cryptocurrenncy is a regression machine learning model

### Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import regression
sns.set()
plt.style.use('seaborn-whitegrid')
```

## Data

> Data gathered from Yahoo Finance from Jan 23, 2019 - Jan 23, 2022

```python
data = pd.read_csv('DOGE-USD.csv')
```

```python
data.head()
```

> This dataset the close column contains the values whose future values we want to predict. Let's have a close look at them:

```python
data.dropna()
plt.figure(figsize=(10,4))
plt.title("dogecoin price")
plt.xlabel("Date")
plt.ylabel("Close")
plt.plot(data["Close"])
plt.show()
```

> autots library in Python will be used to train a machine learning model for predicting the future prices of dogecoin.

```python
from autots import AutoTS
model = AutoTS(forecast_length=10, frequency='infer',
              ensemble='simple', drop_data_older_than_periods=600)
model = model.fit(data, date_col='Date', value_col='Close', 
                  id_col=None)
prediction = model.predict()
fore = prediction.forecast
print("dogecoin price prediction")
print(fore)
```
